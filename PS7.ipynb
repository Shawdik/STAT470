For this problem, the function for females is l<sub>x</sub> = 1000(1 - <sup>x</sup>/<sub>105</sub>) for 0 ≤ x ≤ 105, and the function for males is l<sub>x</sub> = 1000(1 - <sup>x</sup>/<sub>99</sub>) for 0 ≤ x ≤ 99. 

- Calculate the present value of an annual life annuity-due
- parameters for annual amount, current age (x) , gender (M/F), interest rate effective annual (i)
- first payment made at age x

```python
# Shawdi Keshavarz
# summation from t = 0 to w-x (tPx)(v**t)
def prob1(p:int, x:int, gen:str, i:float):
    answer = 0
    den = 0
    if gen == "M":
        den=99
        # Shawdi Keshavarz
    elif gen == "F":
        den=105
        
    upper_limit=den-x
    v= 1/(1+i)
    
    summation = sum([( (1 - (element/upper_limit)) * (v**element) ) for element in range(0,upper_limit+1)])
    # Shawdi Keshavarz
    answer = round((p*summation),2)
    print(answer)
    
    
prob1(50000, 30, "F", .075)    # should be 589,821.03
prob1(50000, 30, "M", .075)    # should be 579,122.73
prob1(1000, 70, "M", .05)
```
