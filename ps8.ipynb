- Calculate the present value of whole life insurance policy starting at age x
- parameters for face value, current age (x) , gender (M/F), interest rate effective annual (i)
- first payment made at age x

```python
# Shawdi Keshavarz
def prob1(p:int, x:int, gen:str, i:float):
    den = 0
    if gen == "M":
        den = 99
        # Shawdi Keshavarz
    elif gen == "F":
        den = 105
    else:
        den = 115
        
    n = den-x
    v = 1 / (1 + i)
        # Shawdi Keshavarz
    numer = 1 - (v**n)
    denom = 1 - v
    answer = (p/n)*v*(numer/denom)
    # Shawdi Keshavarz
    answer = '${:,.2f}'.format(answer)
    print(answer)
    
    
prob1(100000,50,'F', .08) # should equal 22,397.48 - example 9.2 from book
prob1(100000,40,'',.13) # should equal 10,255.34 - 9.1.b - added just for purpose of checking
```
