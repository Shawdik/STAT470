For this problem, the function for females is l<sub>x</sub> = 1000(1 - <sup>x</sup>/<sub>105</sub>) for 0 ≤ x ≤ 105, and the function for males is l<sub>x</sub> = 1000(1 - <sup>x</sup>/<sub>99</sub>) for 0 ≤ x ≤ 99. 

- Create a function that can **calculate either the curtate or complete expectation of life**. 
- parameters/arguments for age, gender (you can specify male or female), and type of future lifetime (you can specify curtate or complete expectation). 
- Test that your function works by providing a test calculation by hand and comparing it with the value generated by the function.

```python
# Shawdi Keshavarz
import sympy as smp
from sympy import evalf
t=smp.symbols('t', real=True)

# Shawdi Keshavarz
def life(age:int, gender:str, type:str):
    answer=0
    den=0
    # Shawdi Keshavarz
    if gender == "m":
        den=99
    elif gender == "f":
        den=105
        
    upper_limit = den-age
    
    # Shawdi Keshavarz
    if type == "curtate":
        summation=[(element/upper_limit) for element in range(1,(upper_limit+1))]
        answer = upper_limit-sum(summation) 
        
    # Shawdi Keshavarz
    elif type == "complete":
        f = 1-(t/upper_limit)
        answer=(smp.integrate(f, (t,0,upper_limit))).evalf()
    
    return(round(answer,2))

# Shawdi Keshavarz
print(life(90, "m", "curtate"))
print(life(50, "m", "complete"))

print(life(30, "f", "curtate"))
print(life(50, "f", "complete"))
```
